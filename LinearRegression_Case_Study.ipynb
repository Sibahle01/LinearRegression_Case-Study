**Identifying Missing Values**

```python
missing_values = data.isnull().sum()
missing_percentage = (missing_values/len(data)) * 100
print("Missing Values in Each Columns:\n", missing_values)
print("\nPercentage of Missing Data:\n", missing_percentage)
```

**Remove Missing Values**

```python
data_cleaned = data.dropna()
```

**Data Explotation and Visualisation**

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(data_cleaned['median_house_value'], color='forestgreen', kde=True)
plt.title('Distribution of Median House Values')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.show()
```

**Using InterQualtile Range to Remove Outliers**

```python
# Assuming 'data_cleaned' is your dataframe and 'median_house_value' is the column of interest
Q1 = data_cleaned['median_house_value'].quantile(0.25)
Q3 = data_cleaned['median_house_value'].quantile(0.75)
IQR = Q3 - Q1

# Define the bounds for the outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
data_no_outliers = data_cleaned[(data_cleaned['median_house_value'] >= lower_bound) &
                                (data_cleaned['median_house_value'] <= upper_bound)]

# Check the shape of the data before and after removal of outliers
print("Original Data shape:", data_cleaned.shape)
print("New data shape without outliers:", data_no_outliers.shape)
```

**BoxPlot for Outlier Detection**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=data_no_outliers['median_income'], color='purple')
plt.title('Outlier Analysis in Median Income')
plt.xlabel('Median Income')
plt.show()
```
